```python
from ultralytics import YOLO
import cv2

# Load a pre-trained YOLO model
model = YOLO("/content/ultralytics/runs/detect/train2/weights/best.pt")

# Open the video file
video_path = "/content/Blackbuck4.MOV"
cap = cv2.VideoCapture(video_path)

# Get video properties
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for output file (mp4 format)
output_path = "output_video_with_conf2.mp4"
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Set confidence threshold
conf_threshold = 0.18  # Confidence threshold (adjust as needed)

# Process the video frame by frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform object detection with confidence threshold
    results = model(frame, conf=conf_threshold)

    # Annotate the frame with detection results
    annotated_frame = results[0].plot()

    # Resize labels if needed (adjusting font size)
    for detection in results[0].boxes:
        bbox = detection.xyxy[0].cpu().numpy()  # Bounding box
        label = detection.cls  # Class label
        conf = detection.conf  # Confidence score

        # Modify text scale and thickness for smaller labels
        label_text = f'{model.names[int(label)]} {conf.item():.2f}'
        font_scale = 0  # Reduce the font size (adjust as needed)
        font_thickness = 1  # Reduce the thickness

        # Draw label text with smaller font size
        cv2.putText(annotated_frame, label_text, (int(bbox[0]), int(bbox[1]) - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, font_scale, (255, 255, 255), font_thickness, cv2.LINE_AA)

    # Write the annotated frame to the output video
    out.write(annotated_frame)

# Release resources
cap.release()
out.release()

print(f"Video with detections saved to {output_path}")
```

```python
import os
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Assume model is already trained and loaded
model = YOLO('blackbuck_detector.pt')

# Validation
dataset_path = '/content/ultralytics/blackbuck-detection-3'
validation_results = model.val(data=os.path.join(dataset_path, "data.yaml"))

# Extract validation accuracy
val_accuracy = validation_results.results_dict['metrics/mAP50-95(B)']

# Test
test_data_path = "/content/ultralytics/blackbuck-detection-3/test/images"
test_results = model(test_data_path)

# Calculate test accuracy (mAP50-95)
# Note: This is a simplified calculation and may not be exactly equivalent to the validation mAP
test_accuracies = []
for r in test_results:
    if r.boxes is not None and len(r.boxes) > 0:
        test_accuracies.append(r.boxes.conf.mean().item())  # Use mean confidence as a proxy for accuracy
    else:
        test_accuracies.append(0)  # No detections

test_accuracy = sum(test_accuracies) / len(test_accuracies) if test_accuracies else 0

# Plot accuracies
plt.figure(figsize=(10, 6))
plt.bar(['Validation', 'Test'], [val_accuracy, test_accuracy])
plt.title('Validation vs Test Accuracy')
plt.ylabel('mAP50-95 (Validation) / Mean Confidence (Test)')
plt.savefig('accuracy_comparison.png')
plt.show()

# Save the model
model.save('blackbuck_detector.pt')

print(f"Validation Accuracy (mAP50-95): {val_accuracy}")
print(f"Test Accuracy (Mean Confidence): {test_accuracy}")
```
